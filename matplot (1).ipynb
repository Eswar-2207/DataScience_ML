```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.array([2,4,6,8,9])
b=np.array([10,30,50,70,80])
plt.plot(a,b)
plt.title("days vs sales")
plt.xlabel("days")
plt.ylabel("sales")
plt.show()
```

```python
#a=np.array([2,4,6,8,9])
a=np.array([60,65,70,75,80,85,90,95])
plt.hist(a)
plt.title("days vs sales")
plt.xlabel("days")
plt.ylabel("sales")
plt.show()
```

```python
region=["north","west","south","east"]
sales=[35,55,65,75]
plt.pie(sales,labels=region,autopct='%1.1f%%')
plt.legend()
plt.show()
```

```python
#box plot
d=[10,12,15,18,20,22,25,100]
plt.boxplot(d)
plt.show
```
